```python
import pymupdf
import os

pdf_file_path = 'CH4/data/Thesis_for_Degree_of_Master.pdf'
doc = pymupdf.open(pdf_file_path)
```

```python
doc
```

```python
full_text = ''

for page in doc:
    text = page.get_text()
    full_text += text

pdf_file_name = os.path.basename(pdf_file_path)
```

```python
pdf_file_name = os.path.splitext(pdf_file_name)[0]
```

```python
pdf_file_name
```

```python
txt_file_path = f'CH4/output/{pdf_file_name}.txt'
```

```python
with open(txt_file_path, 'w', encoding='utf-8') as f:
    f.write(full_text)
```
