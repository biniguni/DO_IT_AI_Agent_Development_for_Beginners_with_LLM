```python
import google.genai as genai
from google.genai import types
```

```python
## 싱글턴

GOOGLE_API_KEY = ''

## api_key 세팅
# genai.configure(api_key=GOOGLE_API_KEY)
client = genai.Client(api_key=GOOGLE_API_KEY)

## 프롬프트
prompt = """
2022년 월드컵 우승팀은 어디임?
"""   

## 모델
# model = genai.GenerativeModel('gemini-flash-latest')
# model = genai.GenerativeModel('gemini-2.5-flash')
# model = genai.GenerativeModel('gemini-3-flash-preview')

## client.models module exposes model inferencing and model getters. 모델 추론 및 모델 가져오기 기능?
## 텍스트 컨텐츠 포함
response = client.models.generate_content(
    model='gemini-2.5-flash', 
    contents = prompt
)


## 메시지 전송 및 응답 확인
print(response.text)
```

```python
# JSON schema -> 구조화된 출력

response = client.models.generate_content(
    model='gemini-2.5-flash',
    contents='2022년 월드컵 정보 알려줘',
    config=types.GenerateContentConfig(
        response_mime_type='application/json',
        response_json_schema={
            'required': [
                '개최국',
                '우승팀',
                '준우승팀',
                '결승전 스코어',
                '월드컵 MVP',
                '득점왕',
                
            ],
            'properties': {
                '개최국': {'type': 'STRING'},
                '우승팀': {'type': 'STRING'},
                '준우승팀': {'type': 'STRING'},
                '결승전 스코어': {'type': 'STRING'},
                '월드컵 MVP': {'type': 'STRING'},
                '득점왕': {'type': 'STRING'},
               
            },
            'type': 'OBJECT',
        },
    ),
)
print(response.text)
```

```python
response = client.models.count_tokens(
    model='gemini-2.5-flash',
    contents='2022년 월드컵 우승팀은 어디임?',
)
print(response)
```
