```python
from google import genai
from google.genai import types
from dotenv  import load_dotenv
import os 
```

```python
load_dotenv()
GOOGLE_API_KEY = os.getenv('GEMINI_API_KEY')

client = genai.Client(api_key=GOOGLE_API_KEY)
```

```python
system_instruction = "너는 백설공주 이야기 속의 마법 거울이야. 그 이야기의 캐릭터에 부합하게 답변해줘"

response = client.models.generate_content(
    model='gemini-2.5-flash', 
    # 출력의 무작위성 제어, [0.0, maxTemperature], 값이 클수록 다양한 대답? 0에 가까울수록 덜 놀라운 대답?
    contents=['세상에서 누가 제일 예쁘니?'],
    config=types.GenerateContentConfig(
        system_instruction=system_instruction,
        temperature=0.9    
    )
)
```

```python
print(response.text)
```

```python
system_instruction = '너는 배트맨에 나오는 배트맨이야. 배트맨 캐릭터에 맞게 답변해줘.'



response = client.models.generate_content(
    model='gemini-2.5-flash',
    contents='세상에서 누가 제일 예쁘니?',
    config=types.GenerateContentConfig(
        system_instruction=system_instruction,
        temperature=0.9    
    )
)
```
