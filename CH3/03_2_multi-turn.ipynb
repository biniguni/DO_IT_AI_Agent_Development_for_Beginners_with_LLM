```python
## multi-turn : 여러 번 대화(턴) 할 때 이전 대화를 기억하고 적절하게 반응하는 것
from google import genai
from google.genai import types
from dotenv import load_dotenv
import os

load_dotenv()
GOOGLE_API_KEY = os.getenv('GEMINI_API_KEY')

client = genai.Client(api_key=GOOGLE_API_KEY)
```

```python
## 이전 대화 내용 기억 못함

while True:
    user_input = input("user : ")
    
    if user_input == 'exit':
        break
    
    respones = client.models.generate_content(
        model='gemini-2.5-flash',
        contents=[
            {'role':'user', 'parts':[{'text': user_input}]}
        ],
        config=types.GenerateContentConfig(
            temperature=0.1,
            system_instruction='너는 사용자를 도와주는 상담사야'            
        )
    )
    print('model:' + respones.text)
```

```python
def get_ai_response(messages):
    chat = client.chats.create(model='gemini-2.5-flash')
    response = chat.send_message(messages)
    return response.text

messages = [
    {'role':'user', 'parts': [{'text' : '너는 사용자를 도와주는 상담사야'}]}
]
```

```python
while True:
    user_input = input('사용자: ')
    
    if user_input == 'exit':
        break
    
    messages.append({[
    {'role':'user', 'parts': [{'text' : user_input}]}
]})
    ai_response = get_ai_response(messages)
    messages.append({[
    {'role':'model', 'parts': [{'text' : ai_response}]}
]})
    print('AI: ') + ai_response
```

```python
chat = client.chats.create(
    model='gemini-2.5-flash',
    config=types.GenerateContentConfig(
        system_instruction="너는 사용자를 도와주는 상담사야."
    )
)

while True:
    user_input = input('사용자: ')
    if user_input == 'exit':
        break
    
    # 2. 메시지 전송 (chat 객체가 이전 대화를 기억함)
    response = chat.send_message(user_input)
    
    print(f'AI: {response.text}')
```
